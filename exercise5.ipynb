```r
### Statistical Modeling 2
### Functional Data Models

############################
### Assignment 5

### 1
library(fda)
data("growth")
Y1 <- growth$hgtf # girls
Y2 <- growth$hgtm # boys
t <- growth$age


par(mfrow=c(1,2))
matplot(t, Y1, pch=19, col="red", main="Height - Girls")
matlines(t, Y1, col="red")
matplot(t, Y2, pch=19, col="blue", main="Height - Boys")
matlines(t, Y1, col="blue")

head(Y1[,1:5])
head(Y2[,1:5])
```

```r
### a)

## B-spline basis

basis.g <- create.bspline.basis(c(1, 18), nbasis = 15, norder = 6)
plot(basis.g)
Y1.fd <- smooth.basis(t, Y1, basis.g) # least squares method
plot(Y1.fd)

# coeffiecients
coef.g <- Y1.fd$fd$coefs
coef.g

# fitted values
fitted.g <- eval.fd(t, Y1.fd$fd)
plot(fitted.g[,1], Y1[,1])
```

```r
### b)

basis.g <- create.bspline.basis(c(1, 18), nbasis = 20, norder = 6)
plot(basis.g)
basisPar.g <- fdPar(basis.g, Lfd = 2, lambda = 0.01) # penalized least squares method
Y1.fd <- smooth.basis(t, Y1, basisPar.g)
plot(Y1.fd, col="red")

coef.g <- Y1.fd$fd$coefs
coef.g

fitted.g <- eval.fd(t, Y1.fd$fd)
plot(fitted.g[,1], Y1[,1])
```

```r
### c)

basis.b <- create.bspline.basis(c(1, 18), nbasis = 20, norder = 6)
basisPar.b <- fdPar(basis.b, Lfd = 2, lambda = 0.01) # penalized least squares method
Y2.fd <- smooth.basis(t, Y2, basisPar.b)
plot(Y2.fd)

coef.b <- Y2.fd$fd$coefs
coef.b

fitted.b <- eval.fd(t, Y2.fd$fd)
plot(fitted.b[,1], Y2[,1])

mean.Y1fd <- mean(Y1.fd$fd) # sample mean functions for girls
mean.Y2fd <- mean(Y2.fd$fd) # sample mean functions for boys

par(mfrow=c(1,2))
plot(Y1.fd)
lines(mean.Y1fd, col="red", lwd=4)
plot(Y2.fd)
lines(mean.Y2fd, col="blue", lwd=4)

eval.fd(15, mean.Y1fd) # sample mean for girls at t = 15years
# 165

eval.fd(15, mean.Y2fd) # sample mean for boys at t = 15years
# 174
```

```r
### d)

D.Y1 <- eval.fd(t, Y1.fd$fd, 1) # First Derivates
dim(D.Y1)
# [1] 31 54

plot(t, seq(min(D.Y1), max(D.Y1), length=31), type="n", main="Derivates") 
for(j in 1:dim(D.Y1)[2]){
  lines(t, D.Y1[,j], col="orange", lwd=2)
}

# approximate time points when the first derivatives is zero
# 5.5 and 11.25
```

```r
### e

basis.f<-create.bspline.basis(c(1, 19), nbasis=20, norder=6) # including 19 now
plot(basis.f)
basisPar.f <- fdPar(basis.f, Lfd = 2, lambda = 0.01) # penalized least squares method
Y1.f <- smooth.basis(t, Y1, basisPar.f) # smooth Y values with the new basis to include the 120 years
plot(Y1.f)

t.f <- t
t.f[32] <- 19
phi.f <- eval.basis(t.f, basis.f)
c.hat <- Y1.f$y2cMap%*%Y1 # extract the c coefficients
pred <- phi.f%*%c.hat # predictions from 1 t0 19

# predictions from 18 to 19 years
pred.1 <- pred[31:32, 1] # 159 159

plot(t.f, seq(min(pred), max(pred), length=32), type="n")
lines(Y1.f, col="red")

lines(c(18, 19), c(pred.1[1], pred.1[2]), col="blue", lwd=3)


# for(i in 1:dim(Y1)[2]){
#   
#   lines(t.f, pred[,i], col="brown", lwd=3)
#   
# }

```
