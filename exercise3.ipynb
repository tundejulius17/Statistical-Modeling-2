```r
### e)

## prediction interval for other elements of the random vector y given some other elements of the vector
yf1<-c(148.7, 156.6, 164.7)
Bt<-t(coef(model.G))
B1t<-Bt[1:3,]
B2t<-Bt[4:9,]
Sigma.G <- Sigma
Sigma11.G<-Sigma.G[1:3,1:3]
Sigma12.G<-Sigma.G[1:3,4:9]
Sigma21.G<-Sigma.G[4:9,1:3]
Sigma22.G<-Sigma.G[4:9,4:9]

newdata<-data.frame(gender = 1)
xf.G <- t(cbind(1, newdata))
BP<-B2t%*%xf.G + Sigma21.G%*%solve(Sigma11.G)%*%(yf1-B1t%*%xf.G) # point prediction
# [,1]
# Y13 170.1792
# Y14 172.7627
# Y15 173.1339
# Y16 173.0139
# Y17 172.9713
# Y18 173.0809

cov.error<-Sigma22.G-Sigma21.G%*%solve(Sigma11.G)%*%Sigma12.G # estimated covariance matrix of the prediction error

lower.bound<-BP-qnorm(0.9)*diag(cov.error)^0.5
upper.bound<-BP+qnorm(0.9)*diag(cov.error)^0.5
lower.bound
upper.bound


# > lower.bound
# [,1]
# Y13 168.4604
# Y14 169.2501
# Y15 169.0688
# Y16 169.0881
# Y17 168.9461
# Y18 168.8734
# > upper.bound
# [,1]
# Y13 171.8980
# Y14 176.2752
# Y15 177.1991
# Y16 176.9396
# Y17 176.9966
# Y18 177.2884
```

```r
### 2

excretory <- read.table("excretory.txt", sep="\t", dec=".", header=TRUE)
attach(excretory)
summary(excretory)

# gravity           obesity     creatinine      chloride         chlorine     
# Min.   :10.00   Group I  :12   Min.   : 8.7   Min.   : 1.550   Min.   : 0.200  
# 1st Qu.:23.00   Group II :14   1st Qu.:12.9   1st Qu.: 4.050   1st Qu.: 3.000  
# Median :26.00   Group III:11   Median :15.0   Median : 5.100   Median : 5.200  
# Mean   :25.82   Group IV : 8   Mean   :15.9   Mean   : 5.247   Mean   : 7.467  
# 3rd Qu.:31.00                  3rd Qu.:18.5   3rd Qu.: 6.300   3rd Qu.:10.500  
# Max.   :35.00                  Max.   :23.7   Max.   :10.250   Max.   :22.500

Y <- cbind(creatinine, chloride, chlorine)

model.E <- lm(Y ~ gravity + factor(obesity))
summary(model.E)
```

```r
### a)

newdata <- data.frame(gravity=35, obesity="Group III")
BP <- predict(model.E, newdata=newdata) # point prediction
BP
#     creatinine chloride chlorine
# 1   15.05839 7.215861 11.81878

### prediction interval the whole elements of the random vector y
X <- model.matrix(model.E)
xf <- t(cbind(1, 35, 0, 1, 0))
E <- residuals(model.E)
Sigma <- (t(E)%*%E)/model.E$df
Sigma
cov.error <- Sigma*as.numeric(1+t(xf)%*%solve(t(X)%*%X)%*%xf) 

lower.bound <- BP-qnorm(0.9)*diag(cov.error)^0.5
upper.bound <- BP+qnorm(0.9)*diag(cov.error)^0.5
lower.bound
upper.bound

# > lower.bound
#     creatinine chloride chlorine
# 1   10.40143 5.232236 4.394969
# > upper.bound
#     creatinine chloride chlorine
# 1   19.71535 9.199485 19.24258
```

```r
### b)

model.H1 <- lm(cbind(creatinine, chloride) ~ gravity + obesity)
model.H0 <-lm(cbind(creatinine, chloride) ~ obesity)
test <- anova(model.H0, model.H1)
test$`approx F`[2] # value of test statistic = [1] 21.42967
test$`Pr(>F)`[2]  # p-value = 5.259306e-07
```

```r
### 3a)

Y.G <- cbind(Y16, Y17, Y18)
L1.G <- Y17 - Y16
L2.G <- Y18 - Y17
YL.G <- cbind(L1.G, L2.G)

model.YL<- lm(YL.G ~ factor(gender))
summary(model.YL)


library(MASS)

K <- t(c(1,1))
T <- diag(2)-ginv(K)%*%K

X <- model.matrix(model.YL)
X.star <- X%*%T

model.H0 <- lm(YL.G ~ X.star-1)
summary(model.H0)
test <- anova(model.H0,model.YL) 

test$`approx F`[2] # 10.01203
test$`Pr(>F)`[2]  # 0.0001185677
```
